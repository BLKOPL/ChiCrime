```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import gmaps
import pygeocoder as Geocoder
import geocoder
import json
import pprint


# Google developer API key

#from config import gkey

# Configure gmaps


# File to Load
crimes_data = "Input/crime_data.csv"

# Read the Crimes Data
crimes_data = pd.read_csv(crimes_data)


# Display the data table for preview
crimes_data.head()
```

```python
location_df = pd.DataFrame(crimes_data)
cols = [-1,-3,-2,]
location_df = location_df[location_df.columns[cols]]
location_df.head(10)
```

```python
# 1.Convert longitude and latitude to a location
#results = Geocoder.reverse_geocode(location_df['latitude'][0], location_df['longitude'][0])

results = geocoder.google([location_df['Latitude'][0], location_df['Longitude'][0]], method='reverse')  # lat and lon you've extracted already, but maybe you'll still need to cast them to float.

resp = results.geojson

#json.loads(resp)

resp
```
